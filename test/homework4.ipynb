**Exercise 3.** The company wants to find the “demand curve”: how many it will sell at each
potential price. Let $P$ denote price and $S$ denote sales per week. Find the line
$S = c_1 + c_2 P$ that best fits the data from the table in the sense of least squares. Find the
normal equations and the coefficients $c_1$ and $c_2$ of the least squares line. Plot the least
squares line along with the data, and calculate the root mean square error.

```julia
A = [1 0.59; 1 0.80; 1 0.95; 1 0.45; 1 0.79; 1 0.99; 1 0.90; 1 0.65; 
    1 0.79; 1 0.69; 1 0.79; 1 0.49; 1 1.09; 1 .95; 1 .79; 1 .65; 1 .45; 
    1 .60; 1 .89; 1 .79; 1 .99; 1 .85]
b = [3980; 2200; 1850; 6100; 2100; 1700; 2000; 4200; 2440; 3300; 2300; 6000; 1190; 1960; 
2760; 4330; 6960; 4160; 1990; 2860; 1920; 2160]

x = LeastSquares.evaluateNormal(A, b)

LeastSquares.least_squares_na(A, b)

# Extract the intercept (x[1]) and slope (x[2]) from the solution vector
intercept = x[1]
slope = x[2]

# Plot the data points with labels
plot(A[:, 2], b, seriestype = :scatter, label = "Data Points",
    title = "Demand Curve for Price vs. Sales (per Week)",
    xlabel = "Price",
    ylabel = "Sales (per Week)")

# Generate the predicted points using the intercept and slope
x_values = A[:, 2]
y_values = intercept .+ slope .* x_values

# Plot the line of best fit
plot!(x_values, y_values, label = "Least Squares Regression Fit")
```
