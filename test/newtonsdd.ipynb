```julia
#Example Usage 

x = [0.0, 1.0, 2.0, 3.0, 4.0]
y = [1.0, 2.0, 3.0, 6.0, 11.0]
xᵢ= 0:0.1:4
yᵢ= []
for i in eachindex(xᵢ)
    push!(yᵢ, newton_interpolation(x,y,xᵢ[i]))
end

for i in 1:length(xᵢ)
    println("$(xᵢ[i]) : $(yᵢ[i])")
end

plot(xᵢ, yᵢ, label="Newtons Divided Difference Approximation", xlabel="x", ylabel="y")
```
